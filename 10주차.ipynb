```python
import tiktoken
enc = tiktoken.get_encoding('cl100k_base')
```

```python
text_en = "I like milk"
tokens_en = enc.encode(text_en)
print("нҶ нҒ° мҲҳ:", len(tokens_en))
print("нҶ нҒ°  ID:", tokens_en)
```

```python
text_ko = "лӮҳлҠ” мҡ°мң к°Җ мўӢм•„"
tokens_ko = enc.encode(text_ko)
print("нҶ нҒ° мҲҳ:", len(tokens_ko))
print("нҶ нҒ°  ID:", tokens_ko)
```

```python
from konlpy.tag import Okt
okt=Okt()
print(okt.morphs("мҡ°мң  л§ҲмӢңкі  мӢ¶лӢӨ."))
print(okt.pos("мҡ°мң  л§ҲмӢңкі  мӢ¶лӢӨ."))
```
